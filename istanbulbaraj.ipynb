date : for time series
date2 : for displaying each year in same timespan

```python
date=[]
date2=[]
```

```python
for i in range(len(tarih)):
    try:
        
        date.append(datetime.strptime(tarih[i], "%Y-%m-%d"))
    
        date2.append(datetime.strptime(tarih[i][5:], "%m-%d"))
    except ValueError:
        date2.append(datetime.strptime(tarih[i][5:][:-1]+"8", "%m-%d"))
    


        
```

```python
#plt.figure(dpi=500,figsize=(12,8),facecolor='w', edgecolor='k')
#plt.figure(dpi=500,figsize=(12,8),facecolor='w', edgecolor='k')
fig,ax1 = plt.subplots()
fig.set_dpi(500)


for i in range(len(dateindex)-1):
    
    plt.plot(date2[0:365],dolulukoran[dateindex[i]:dateindex[i+1]],label="{yil}".format(yil=2005+i))
    
    print(i)
    

 
monthyearFmt = mdates.DateFormatter('%B')
ax1.xaxis.set_major_formatter(monthyearFmt)    
plt.plot(date2[-326:],dolulukoran[-326:],label="{yil}".format(yil=2020))
plt.xticks(rotation=90)

plt.xlabel("Tarih")
plt.legend(title='YÄ±llar', bbox_to_anchor=(1.05, 1), loc='upper left', fontsize='small')
plt.savefig('su.png')

#plt.legend(loc ='best')
```
